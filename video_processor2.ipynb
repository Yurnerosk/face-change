```python
import insightface
from insightface.app import FaceAnalysis
from insightface.data import get_image as ins_get_image
```

```python
app = FaceAnalysis(name='buffalo_l')
app.prepare(ctx_id=0, det_size=(640,640))
```

```python
import numpy as np
import glob
import matplotlib.pyplot as plt

import cv2
import os
from IPython.display import Video, display

# G
img_swap = cv2.imread('eehhh.jpg')
g_face = app.get(img_swap)[0]

# Daigo
img_swap = cv2.imread('daigo.jpeg')
daigo_face = app.get(img_swap)[0]

# takumi
img_swap = cv2.imread('takumi2.png')
takumi_face = app.get(img_swap)[0]

# mogi
img_swap = cv2.imread('mogi.png')
mogi_face = app.get(img_swap)[0]
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
swapper = insightface.model_zoo.get_model('inswapper_128.onnx',
                                          download=False,
                                          download_zip=False)
```

```python
# --- Configurações ---
input_video_path = 'videoplayback.mp4'  # Substitua pelo caminho do seu vídeo de entrada
output_video_path = 'output_video.mp4' # Caminho para o vídeo de saída

# Certifique-se de que o vídeo de entrada existe
if not os.path.exists(input_video_path):
    print(f"Erro: Vídeo de entrada não encontrado em '{input_video_path}'")
    print("Por favor, substitua 'input_video.mp4' pelo caminho correto do seu vídeo.")
else:
    # --- Carregar o vídeo ---
    cap = cv2.VideoCapture(input_video_path)

    if not cap.isOpened():
        print(f"Erro ao abrir o vídeo: {input_video_path}")
    else:
        # --- Obter propriedades do vídeo ---
        frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
        # Obter o FPS do vídeo de entrada.
        # Você pode substituir 'output_fps' por um valor fixo se desejar um FPS diferente para o vídeo de saída.
        input_fps = cap.get(cv2.CAP_PROP_FPS)
        output_fps = int(input_fps) # Usar o FPS do vídeo de entrada por padrão
        # Exemplo para definir um FPS fixo de 30:
        # output_fps = 30 
        
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        print(f"Processando vídeo: {input_video_path}")
        print(f"Resolução: {frame_width}x{frame_height}, FPS de entrada: {input_fps}, FPS de saída: {output_fps}, Total de frames: {total_frames}")

        # --- Configurar o gravador de vídeo de saída ---
        # Codec para o vídeo de saída (ex: 'mp4v' para .mp4, 'XVID' para .avi)
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(output_video_path, fourcc, output_fps, (frame_width, frame_height))

        if not out.isOpened():
            print(f"Erro ao criar o arquivo de vídeo de saída: {output_video_path}")
        else:
            frame_count = 0
            while True:
                ret, frame = cap.read()
                if not ret:
                # if not ret or frame_count==300:
                    break # Fim do vídeo ou erro

                frame_count += 1
                # print(f"Processando frame {frame_count}/{total_frames}...")

                # --- ESPAÇO PARA SUAS MODIFICAÇÕES NO FRAME ---
                # Por exemplo, para converter para escala de cinza:
                # modified_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                # modified_frame = cv2.cvtColor(modified_frame, cv2.COLOR_GRAY2BGR) # Converta de volta para 3 canais se o codec exigir

                # ------ frame annotation

                # # Número que você quer colocar
                # numero = str(frame_count)

                # # Configurações do texto
                # fonte = cv2.FONT_HERSHEY_SIMPLEX
                # escala = 3
                # cor = (0, 0, 255)  # Vermelho (BGR)
                # espessura = 5

                # # Obter dimensões da imagem
                # altura, largura, _ = frame.shape

                # # Obter tamanho do texto
                # (tamanho_texto_largura, tamanho_texto_altura), _ = cv2.getTextSize(numero, fonte, escala, espessura)

                # # Calcular posição central
                # pos_x = (largura - tamanho_texto_largura) // 2
                # pos_y = (altura + tamanho_texto_altura) // 2

                # # Colocar o texto na imagem
                # cv2.putText(frame, numero, (pos_x, pos_y), fonte, escala, cor, espessura)



                # ----- insightface
                faces = app.get(frame)
                print(len(faces))

                modified_frame = frame.copy()
                for face in faces:

                    # more like takumi than mogi?
                    if cosine_similarity(face.embedding, takumi_face.embedding)> cosine_similarity(face.embedding, mogi_face.embedding):
                        modified_frame = swapper.get(modified_frame, face, daigo_face, paste_back=True)
                        print('taku')
                    else:
                        modified_frame = swapper.get(modified_frame, face, g_face, paste_back=True)
                        print('mogi')

                # Ou apenas use o frame original por enquanto:
                # modified_frame = frame
                # --- FIM DO ESPAÇO PARA MODIFICAÇÕES ---

                # Certifique-se de que o frame modificado tenha 3 canais se o original tiver
                if len(modified_frame.shape) == 2: # Se for escala de cinza
                    modified_frame = cv2.cvtColor(modified_frame, cv2.COLOR_GRAY2BGR)


                # --- Salvar o frame modificado no novo vídeo ---
                out.write(modified_frame)
                print(f'{frame_count} de {total_frames}')

            print(f"Processamento concluído. {frame_count} frames processados.")
            print(f"Vídeo de saída salvo em: {output_video_path}")

            # Exibir o vídeo de saída no Jupyter Notebook (se o ambiente permitir)
            # display(Video(output_video_path, embed=True))

        # --- Liberar recursos ---
        cap.release()
        out.release()
```
